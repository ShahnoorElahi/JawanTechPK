```python
# task_6_check_key.py

# The dictionary to check
my_dict = {
    "name": "Sarah",
    "age": 25,
    "city": "Karachi"
}

key_to_check = input("Enter the key you want to check: ")

if key_to_check in my_dict:
    print(f"Yes, the key '{key_to_check}' exists in the dictionary.")
    print(f"Value: {my_dict[key_to_check]}")
else:
    print(f"No, the key '{key_to_check}' does not exist.")
```
