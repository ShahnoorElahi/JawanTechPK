```python
# task_2_check_numeric.py

# Defining a list with mixed data types
my_list = ["Jawan", "Pakistan", "Python", 100, "Assignment"]

has_number = False

# Iterating through the list using a for loop
for item in my_list:
    # Check if the item is an integer or a float
    if isinstance(item, (int, float)):
        has_number = True
        print(f"Numeric value found: {item}")
        # We can break the loop once a number is found, or let it continue to find all
        # break 

if not has_number:
    print("No numeric values found in the list.")
```
