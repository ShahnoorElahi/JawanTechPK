```python
# task_4_sum_dictionary.py

# A dictionary with numeric values
scores = {
    "quiz_1": 85,
    "quiz_2": 90,
    "quiz_3": 88
}

total = 0

# Loop through the values and add them up
for value in scores.values():
    # Only add if the value is a number (good practice)
    if isinstance(value, (int, float)):
        total += value

print(f"The sum of all numeric items is: {total}")
```
