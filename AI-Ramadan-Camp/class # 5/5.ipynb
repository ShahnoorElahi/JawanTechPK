```python
# task_5_find_duplicates.py

# A list with some duplicate values
numbers = [10, 20, 30, 20, 40, 10, 50, 60, 30]
duplicates = []
checked_items = []

for item in numbers:
    if item not in checked_items:
        checked_items.append(item)
    elif item not in duplicates:
        # If we have seen it before, it is a duplicate
        duplicates.append(item)

print("Original List:", numbers)
print("Duplicate values:", duplicates)
```
