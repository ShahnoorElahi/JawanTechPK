```python
# task_1_calculator.py

# Taking inputs from the user
val1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /, **): ")
val2 = float(input("Enter second number: "))

if operator == '+':
    result = val1 + val2
    print(f"Result: {val1} + {val2} = {result}")
elif operator == '-':
    result = val1 - val2
    print(f"Result: {val1} - {val2} = {result}")
elif operator == '*':
    result = val1 * val2
    print(f"Result: {val1} * {val2} = {result}")
elif operator == '/':
    if val2 != 0:
        result = val1 / val2
        print(f"Result: {val1} / {val2} = {result}")
    else:
        print("Error: Division by zero is not allowed.")
elif operator == '**':
    result = val1 ** val2
    print(f"Result: {val1} to the power of {val2} = {result}")
else:
    print("Invalid operator selected.")
```
